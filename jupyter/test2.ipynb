# 测试ABTest模块

```python
class AbClient:
    headers = {'Content-Type': 'application/json'}
    def getVersionId(self, key, scene_id=4, experiment_id=1, default_version_id=2):
        '''
        1： 智能换电， 2： 线下扫街
        '''
        try:
            link = "%s/ab/scenes/%s/experiments/%s/getVersionId?key=%s&defaultVersionId=%s" % (self.url, scene_id, experiment_id, key, default_version_id)
            response = requests.get(link, timeout=1)
            response_str = str(response.content, encoding="utf8")
            response_json = json.loads(response_str)
            if response_json['code'] == 0:
                return response_json['data']['versionId']
        except:
            return default_version_id

    def getVersionIds(self, data, scene_id=4, experiment_id=1):
        '''
        批量产生，最大的batch为 500
        :param data: ['23456', '8765353']
        :param scene_id:
        :param experiment_id:
        :return:  dict
        '''
        try:
            data2 = {"keys": data, "defaultVersionId":1}
            link = "https://aiengine-abexperiment-inner.hellobike.cn/ab/scenes/%s/experiments/%s/getVersionIds" % (scene_id, experiment_id)
            response = requests.post(link, headers=self.headers, data=json.dumps(data2), timeout=1)
            response_json = response.json()
            if response_json['code'] == 0:
                return response_json['data']['keyVersionIdsMap']
        except:
            res_dic = {}
            for bikeno in data:
                res_dic[bikeno] = 1 if int(bikeno[-1]) % 2 == 0 else 2
            return res_dic
```
